## Keras. Лобовая имплементация решения Созыкина на основе 3-го места в Титанике

```python
# !pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, MaxPooling1D, Dropout, LSTM, Bidirectional, SpatialDropout1D
from tensorflow.keras import utils
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras import utils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Максимальное количество слов 
num_words = 50_000
# Максимальная длина одной статьи
max_comment_len = 100   # покрывает почти все статьи текущей выборки.
```

```python
from sklearn.model_selection import train_test_split
```

```python
final = pd.read_csv('lenta_encoded.csv')
```

```python
final = final.drop('Unnamed: 0', 1)
```

```python
y = final[list(final.columns[1:])].values
y
```

```python
len(list(final.columns[1:]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(final.text.loc[:].values, y,
                                                    test_size=0.2, stratify=y, random_state=42)
```

```python
# X_train
# y_test
```

```python
comments = pd.DataFrame(X_train).astype('str')[0]  # токенайзер может обрабатывать только текстовые последовательности
```

```python
# comments
```

```python
tokenizer = Tokenizer(num_words=num_words)
```

```python
tokenizer.fit_on_texts(comments)
```

```python
# tokenizer.word_index
```

#### Преобразуем статьи в числовые последовательности

```python
sequences = tokenizer.texts_to_sequences(comments)
```

```python
index = 1
print(comments[index])
print(sequences[index])
```

```python
tokenizer.word_index['федеральное']
```

#### Ограничиваем максимальную длину обработки текстов

```python
X_train = pad_sequences(sequences, maxlen=max_comment_len)
```

```python
X_train[0], len(X_train[0])
```

```python
model_lstm = Sequential()
model_lstm.add(Embedding(num_words, 87, input_length=max_comment_len))
model_lstm.add(SpatialDropout1D(0.5))
model_lstm.add(LSTM(40, return_sequences=True))
model_lstm.add(LSTM(40))
model_lstm.add(Dense(87, activation='sigmoid'))
```

```python
model_lstm.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy', 'AUC'])
```

```python
model_lstm.summary()
```

#### Создаем callback для сохранения нейронной сети на каждой эпохе 
если качество работы на проверочном наборе данных улучшилась. 
Сеть сохраняется в файл best_model.h5

```python
model_lstm_save_path = 'best_model_lstm_2.h5'
checkpoint_callback_lstm = ModelCheckpoint(model_lstm_save_path, 
                                      monitor='val_accuracy',
                                      save_best_only=True,
                                      verbose=1)
```

```python
history_lstm = model_lstm.fit(X_train, 
                              y_train, 
                              epochs=20,
                              batch_size=128,
                              validation_split=0.2,
                              callbacks=[checkpoint_callback_lstm])
```

```python
plt.plot(history_lstm.history['accuracy'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history_lstm.history['val_accuracy'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
```
