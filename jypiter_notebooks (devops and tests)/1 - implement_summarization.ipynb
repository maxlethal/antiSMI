```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("IlyaGusev/mbart_ru_sum_gazeta")

model = AutoModelForSeq2SeqLM.from_pretrained("IlyaGusev/mbart_ru_sum_gazeta")
```

```python
def summarized_news(article_text):
    """Делает краткое саммари из новости"""
    input_ids = tokenizer(
    [article_text],
    max_length=600,
    truncation=True,
    return_tensors="pt")["input_ids"]
    
    output_ids = model.generate(
    input_ids=input_ids,
    no_repeat_ngram_size=4)[0]
    
    summary = tokenizer.decode(output_ids, skip_special_tokens=True)
    
    return summary  
```

```python
from transformers import T5ForConditionalGeneration
```

```python
model_head_name = "IlyaGusev/rut5_base_headline_gen_telegram"
tokenizer_head = AutoTokenizer.from_pretrained(model_head_name)
model_head = T5ForConditionalGeneration.from_pretrained(model_head_name)
```

```python
def summerized_news_to_title(summary):
    """Делает заголовок из краткой новости"""
    
    input_ids = tokenizer_head(
    [summary],
    max_length=600,
    add_special_tokens=True,
    padding="max_length",
    truncation=True,
    return_tensors="pt")["input_ids"]

    output_ids = model_head.generate(
    input_ids=input_ids)[0]
    
    title = tokenizer_head.decode(output_ids, skip_special_tokens=True)
    
    return title
```
